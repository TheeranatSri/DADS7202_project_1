```python
df_raw_ = pd.read_csv('/content/drive/MyDrive/data.csv')
df_raw_
```

```python
dataset = df_raw_.copy()
```

```python
y = dataset['Bankrupt?']
X = dataset.drop('Bankrupt?',axis=1)
```

```python
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.2, random_state=66)
```

```python
X_temp
```

```python
train_data = pd.merge(X_temp, y_temp, left_index=True, right_index=True)
train_data
```

```python
from random import seed
s = setup(data = train_data, target = 'Bankrupt?', session_id=666,
          remove_multicollinearity = True, feature_selection = True)
```

```python
best_model = compare_models(sort = 'Accuracy')
```

```python
print(best_model)
```

```python
rank_1 = create_model('lightgbm')
```

```python
# same as print(best_model)
print(rank_1)
```

```python
plot_model(rank_1, plot='feature')
```

```python
plot_model(rank_1, plot = 'confusion_matrix')
```

```python
evaluate_model(rank_1)
```

```python
# finalize rf model
final_catboost = finalize_model(rank_1)
final_catboost
```

```python
test_data = pd.merge(X_test, y_test, left_index=True, right_index=True)
test_data
```

```python
xxx = test_data

unseen_predictions = predict_model(final_catboost, data=xxx)
unseen_predictions.head()
```

```python
# check metric on unseen data
from pycaret.utils import check_metric
check_metric(unseen_predictions['Bankrupt?'], unseen_predictions['Label'], metric = 'Accuracy')
```
